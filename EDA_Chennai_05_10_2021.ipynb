```python
#Univariate 
# Measure of Central Tendencies (Mean,Median,Mode, Quantiles)
#Measure of Dispersion (Range, Variance, Standard deviation, Coef of variance, IQR)
```

```python
# Mean = Sum (X1+ x2 + --- XN) / N
```

```python
data_num.head(2)
```

```python
data_num.mean()
```

```python
import scipy.stats as stats
me_tr=stats.trim_mean(data_num,proportiontocut=0.1)
```

```python
res=pd.DataFrame()
```

```python
res['Mean']=data_num.mean()
```

```python
res['Trim_mean']=me_tr
```

```python
res
```

```python
res['Median']=data_num.median()
```

```python
res['4 percentile']=data_num.quantile(0.04)
res['25 percentile']=data_num.quantile(0.25)
res['75 percentile']=data_num.quantile(0.75)
```

```python
res
```

```python
#Measure of Dispersion (DEviation measure)
res['Range']=data_num.max()-data_num.min()
```

```python
# Variance
# sum((Xi - X.mean()^2)) / N

# Standard deviation
# sqrt(sum((Xi - X.mean()^2)) / N)
```

```python
res['Variance']=data_num.var()
res['SD']=data_num.std()
```

```python
res['Coef_variance']=data_num.std()/data_num.mean()
```

```python
res
```

```python
# IQR
res['IQR']=data_num.quantile(0.75)-data_num.quantile(0.25)
```

```python
res
```

```python
# For item weight how many outlier are there?
#i='Item_Visibility'
for i in data_num.columns:
    iqr=data_num[i].quantile(0.75) - data_num[i].quantile(0.25)
    #print('Inter Quntile Range',iqr)
    ub=data_num[i].quantile(0.75) + 1.5*iqr
    lb=data_num[i].quantile(0.25) - 1.5*iqr
    #print('UB,LB',ub,lb)
    print('The number of outliers in ',i, ' is ',len(data_num[((data_num[i]>ub) | (data_num[i]<lb))]))
```
